```python
import pandas as pd
import os
import numpy as np
from pprint import pprint

# ajustando algumas configurações para prints
pd.set_option('display.max_columns', None)        # todas as colunas
pd.set_option('display.max_rows', None)           # todas as linhas
pd.set_option('display.width', None)              # largura automática
pd.set_option('display.max_colwidth', None)       # conteúdo completo das células
pd.set_option('display.expand_frame_repr', False) # evita quebrar dataframe em várias linhas
```

```python
# pegando a pasta de trabalho para poder extrair os dados do dataset
PASTA_TRABALHO = os.path.dirname(os.path.abspath(os.getcwd()))
```

## Lendo e armazenando todas as tabelas dos anos necessários

```python
# lista com os nomes das pastas que contem os datasets
nomes_arquivos = ["2024", "2025", "2026"]

# dicionario que contêm todos os dataframes do respectivo ano no seguinte formato:
"""
    ano1:
        ocorrencia_ano1,
        acorrencia_pessoa_ano1,
        acorrencia_todas_causas_ano1
    ano2:
        ...
"""
dataframes_por_ano = {}

# percorre e salva todos as dataframes dos anos escolhidos
for ano in nomes_arquivos:
    diretorio_dataframe = os.path.join(PASTA_TRABALHO, "dataset", ano)

    # percorre todas as tabelas que existem nesse ano (só existe um arquivo)
    for tabela in os.listdir(diretorio_dataframe):

        caminho = os.path.join(diretorio_dataframe, tabela)
        data_frame = pd.read_csv(caminho, sep=";", encoding="latin1") # adiciona a tabela em questão ao dicionário temporário

    dataframes_por_ano[ano] = data_frame # adiciona a tabela de dados ao respectivo ano
```

## juntando as tabelas distintas dos 3 anos em somente uma tabela

```python
dataset = pd.concat([df for df in dataframes_por_ano.values()])
```

## Alterando os tipos de dados para não ter nunhum tipo de conflito com as transformações e armazenar de forma correta os dados

```python
print(dataset["km"].map(type).value_counts())

# DATA
dataset["data_inversa"] = pd.to_datetime(
    dataset["data_inversa"],
    format="%Y-%m-%d"
)


# HORÁRIO
dataset["horario"] = pd.to_datetime(
    dataset["horario"],
    format="%H:%M:%S"
).dt.time


# VALORES NUMÉRICOS
# trocando a virgula por ponto para poder converter para float
dataset["latitude"] = pd.to_numeric(
    dataset["latitude"].astype(str).str.replace(",", "."),
    errors="coerce"
)

dataset["longitude"] = pd.to_numeric(
    dataset["longitude"].astype(str).str.replace(",", "."),
    errors="coerce"
)

# passando para o tipo numérico
dataset["latitude"] = pd.to_numeric(
    dataset["latitude"],
    errors="coerce"
)

dataset["longitude"] = pd.to_numeric(
    dataset["longitude"],
    errors="coerce"
)

dataset["km"] = pd.to_numeric(
    dataset["km"].astype(str).str.replace(",", "."),
    errors="coerce"
)

dataset["km"] = pd.to_numeric(
    dataset["km"],
    errors="coerce"
)
```

```python
# printando algumas informações pertinentes que ajudam a avaliar quais transformações são necessárias
display(dataset.shape)
display(dataset.head())
display(dataset.info())
display(dataset.describe())
display(dataset.isnull().sum())
```

## Avaliando os valores da coluna idade

Nessa coluna tem dois problemas principais:
- Quando não existe informação de qual a idade ela é preenchida como -1, é melhor fica como NULL pois deixar esse campo com -1 pode atrapalhar o calculo de métricas posteriormente como saber qual a média de idade das pessoas que sofrem acidentes
- Existem idades que são impossíveis como por exemplo maiores que 200 anos, avaliando os dados notem que existem amostras com 127 anos depois com 225 anos, apesar de idades superiores a 110 anos já serem extremamente improváveis e a pessoa mais velha registrada ter 122 anos vou considerar o limite como 127
- um ponto que provavelmente deve ser um erro, mas não tem como saber de forma 100% correta são as pessoas com 0 anos, isso poderia indicar pessoas com menos de 1 ano de idade, entretando como esse valor detem mais de 78.000 amostras enquanto a quantidade e pessoas maiores que 0 e menor ou igual a 25 anos é de aproximadamente 73.000 leva a indicar que provavelmente esses dados foram preenchidos de forma errada

```python
# analisando os valores distintos dessa coluna para saber quando existem prováveis outliers
# display(dataset["idade"].value_counts().sort_index())

# analisando a quantidade de pessoas com idade igual a 0 e comparando com outro grupo
idade_0_a_25 = dataset.loc[(dataset["idade"] <= 25) & (dataset["idade"] > 0), "idade" ].count()
idade_0 = dataset.loc[dataset["idade"] == 0, "idade"].count()
print(f"Quantidade de pessoas com 0 (zero) anos de idade: {idade_0}")
print(f"Quantidade de pessoas com idades entre 1-25 anos: {idade_0_a_25}")
# apesar de ser inprovável que a quantidade de vítimas recem nascidas seja superior as de idade entre 1-25 anos não é possível dizer que isso é um erro de forma categorica

# substituindo os valores -1 por NULL
dataset.loc[dataset["idade"] == -1, "idade"] = pd.NA

# substituindo idade impossíveis
dataset.loc[dataset["idade"] > 127, "idade"] = pd.NA
```

# Tratando a coluna classificacao_acidente
Existem alguns valores nulos nessa coluna que possivelmente podem ser passados para um tipo dentro das categorias já existentes

```python
# De acordo com o dicionário de dados existe a classificação "Ignorado" para essa coluna, vamos passar os valores NULL para essa classe
dataset["classificacao_acidente"] = dataset["classificacao_acidente"].fillna("Ignorado")
```

## Tratando os valores nulos das colunas: tipo_envolvido, stado_fisico e sexo
Como cada uma dessas colunas contem a mesma quantidade de valores nulos pode ser que sejam exatamente a mesma linha que não contem multiplas informações, uma pessoa que não contem nenhuma dessas informações pode ser caracterizada como não identificada para facilitar o entendimento das informações

```python
# verificando se realmente é a mesma linha com multiplas linhas colunas vazia
filtro = (
    dataset["estado_fisico"].isna() &
    dataset["tipo_envolvido"].isna() &
    dataset["sexo"].isna()
)
print(f"quantidade de linhas que não tem informação nas colunas em questão: {filtro.sum()}")

# agora que foi mostrado que são a mesma linha vou avaliar qual a idade descrita, levando em consideração que essa é a ultima informação que sobre referente a pessoa envolvida
print(f"desas linhas quais as idades das pessoas em questão: {dataset.loc[filtro, "idade"].unique()}")

# avaliando quantos pesid distintos existem para essa pessoa
print(f"quantidade de pesid existentes para cada uma dessas linhas: {dataset.loc[filtro, "pesid"].nunique()}")

# avaliando quantos id_veiculo distintos existem para essa pessoa
print(f"quantidade de id_veiculo existentes para cada uma dessas linhas: {dataset.loc[filtro, "id_veiculo"].nunique()}")

# avaliando o tipo do veiculo envolvido pera essas pessoas
print(f"quantidade de veiculos sem tipo definido para essas pessoas: {dataset.loc[filtro, "tipo_veiculo"].isna().sum()}")

# aparentemente quando uma pessoa não é identificada ela é colocada como um ID específico e com os valores nulos nessas respectivas colunas e idade igual a 0 (zero)
# criando uma coluna nova que vai dizer se a pessoa em questão foi ou não identificada para facilitar o entendimento dos dados
dataset["identificada"] = 1
dataset.loc[filtro, "identificada"] = 0
```
